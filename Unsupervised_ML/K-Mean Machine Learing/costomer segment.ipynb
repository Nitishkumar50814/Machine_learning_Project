### Kmean Clustering

```python
data.columns
```

```python
x = data[['Annual Income (k$)','Spending Score (1-100)' ]]
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans()
k_means.fit(x)
```

```python
k_mean = KMeans(n_clusters = 5)
k_means.fit_predict(x)
```

## Elbow method  to find Optimal Number of Clusters

```python
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters=i)
    k_means.fit(x)
    wcss.append(k_means.inertia_)

```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel('Numbers of Clusters')
plt.show()
```

## Model training

```python
x = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
k_means = KMeans(n_clusters = 5, random_state = 42)
y_means = k_means.fit_predict(x)
```

```python
y_means
```

```python
plt.figure(figsize=(16,7))
plt.scatter(x.iloc[y_means==0,0],x.iloc[y_means==0,1],s=100,c='red',label='Clusters 1')
plt.scatter(x.iloc[y_means==1,0],x.iloc[y_means==1,1],s=100,c='blue',label='Clusters 2')
plt.scatter(x.iloc[y_means==2,0],x.iloc[y_means==2,1],s=100,c='yellow',label='Clusters 3')
plt.scatter(x.iloc[y_means==3,0],x.iloc[y_means==3,1],s=100,c='green',label='Clusters 4')
plt.scatter(x.iloc[y_means==4,0],x.iloc[y_means==4,1],s=100,c='black',label='Clusters 5')
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1],s=200,c='magenta',marker='*')

plt.title('Customer Segmentation')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
